```python
data = 'Product_id\tQuantity\tSellPrice\tActualPrice\n'
n = int(input("Enter number of rows : "))
print("\n\tEnter data as ProductId, Quantity, SellPrice, ActualPrice : ")
for var in range(n): 
    d = input().split(',')
    data += '\t'.join(d)+'\n'
print(data)
```

```python
f = open("sells.tsv","w")
f.write(data)
f.close()
print("Data successfully Writen in file sells.tsv")
```

```python
#now find out total_sell and profit on data

fp  = open('sells.tsv')
sells_data = fp.read().split('\n')
sells_data = [ entry.split('\t') for entry in sells_data ]
```

```python
print(sells_data)
```

```python
# price of one product --> quantity*sellprice
# actual costing of product --> quantity*actualprice
#profit = sellprice - actualprice
total_sell = 0
total_profit = [ ] # 
for entry in sells_data[1:] : 
    if entry : 
        continue
    entry[1:] = map(int,entry[1:])
    total_profit.append(entry[1]*(entry[2]-entry[3]))
    total_sell += entry[1]*entry[2]
```

```python
print("Total Sell --> ",total_sell)
```

```python
print("Profit --> ",sum(total_profit))
```
